```python
# Assignment 5.2: Add all numbers divisible by both 7 and 9 from 1 to n

n = int(input("Enter the value of n: "))

total = 0
for i in range(1, n + 1):
    if i % 7 == 0 and i % 9 == 0:
        total += i

print("The sum of numbers divisible by both 7 and 9 from 1 to", n, "is:", total)
```

```python
# Assignment 5.3: Add all prime numbers from 1 to n

n = int(input("Enter the value of n: "))

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

total = 0
for i in range(1, n + 1):
    if is_prime(i):
        total += i

print("The sum of all prime numbers from 1 to", n, "is:", total)
```

```python
# Assignment 6.1: Add all odd numbers from 1 to n using a function

def add_odd_numbers(n):
    total = 0
    for i in range(1, n + 1):
        if i % 2 != 0:
            total += i
    return total

# Input from user
n = int(input("Enter the value of n: "))

# Function call and result display
print("The sum of all odd numbers from 1 to", n, "is:", add_odd_numbers(n))
```

```python
# Assignment 6.2: Add all prime numbers from 1 to n using a function

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def add_prime_numbers(n):
    total = 0
    for i in range(1, n + 1):
        if is_prime(i):
            total += i
    return total

# Input from user
n = int(input("Enter the value of n: "))

# Function call and result display
print("The sum of all prime numbers from 1 to", n, "is:", add_prime_numbers(n))
```
