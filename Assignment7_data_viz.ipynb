```python
# ---------------------------------------------------------
# Assignment 7: Na誰ve Bayes Classifier (Using sklearn)
# ---------------------------------------------------------

# Step 1: Import necessary libraries
from sklearn.datasets import load_iris       # Example dataset
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# ---------------------------------------------------------
# Step 2: Load the dataset
# ---------------------------------------------------------
iris = load_iris()
X = iris.data        # Features
y = iris.target      # Target labels

# Optional: Convert to DataFrame for better readability
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y

print("Dataset Loaded Successfully!")
print(df.head(), "\n")

# ---------------------------------------------------------
# Step 3: Split the dataset into training and testing sets
# ---------------------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# ---------------------------------------------------------
# Step 4: Create and train the Gaussian Na誰ve Bayes model
# ---------------------------------------------------------
model = GaussianNB()
model.fit(X_train, y_train)

# ---------------------------------------------------------
# Step 5: Predict the output for the test data
# ---------------------------------------------------------
y_pred = model.predict(X_test)

# ---------------------------------------------------------
# Step 6: Evaluate the model
# ---------------------------------------------------------
accuracy = accuracy_score(y_test, y_pred)
conf_mat = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred, target_names=iris.target_names)

print("------------------------------------------------------")
print("Na誰ve Bayes Classifier Evaluation Results")
print("------------------------------------------------------")
print(f"Accuracy Score: {accuracy * 100:.2f}%\n")
print("Confusion Matrix:")
print(conf_mat)
print("\nClassification Report:")
print(report)
print("------------------------------------------------------")
```
